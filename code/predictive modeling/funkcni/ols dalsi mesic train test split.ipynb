```python

import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

df = pd.read_excel(fr"C:\Users\halik\OneDrive\development\github\sws\code\predictive modeling\funkcni\funcni data set.xlsx")
x_new = df
df = df.drop(df.index[:13])
df = df.drop(df.index[-12:])

datum = df['datum']
x = df[["index", 'repo','repo-1', 'aktivni-1','nove-1','doba-1']]

y = df[['aktivni']]

x_test = x[70:]
x = x[:70]
y_test = y[70:]
y = y[:70]


#fit linear regression model
model = sm.OLS(y,x).fit()

# Predict on the test set
y_pred = model.predict(x_test)
print(y_pred)
# Evaluate the best model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)


print(f"Mean absolute error: {mae:.2f}")
print(f"Mean squared error: {mse:.2f}")
print(f"Root mean squared error: {rmse:.2f}")

#view model summary
print(model.summary())

graf =True
if graf:

    plt.plot(y_pred, label='y_pred')
    #plt.plot(y_train.values, label='y_train')
    plt.plot(y_test, label='y_test')
    plt.legend()
    plt.show()

```
